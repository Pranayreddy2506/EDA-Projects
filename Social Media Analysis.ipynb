```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\reddy\OneDrive\Documents\dummy_data.csv")
data.head()
```

```python
#Age below or equal to 18 using social media
data1 = data[data['age'] <= 18]
data1
```

```python
# Count of locations
country_counts = data['location'].value_counts()
print(country_counts)

# Count of Gender
gender_count = data['gender'].value_counts()
print(gender_count)

# Count of Platform using
platform_count = data['platform'].value_counts()
print(platform_count)
```

```python
data[data['profession'] == 'Student']
```

```python
Using_count = data['profession'].value_counts()
# Plot the pie chart
plt.figure(figsize=(4, 4))
plt.pie(Using_count, labels=Using_count.index, autopct='%1.1f%%', colors=sns.color_palette('muted'), startangle=150)
plt.title('Distribution of using platform')
plt.show()
```

```python
grouped_data = data.groupby(['platform', 'age']).size().unstack()

# Create the bar chart
grouped_data.plot(kind='bar', stacked=True)
plt.xlabel('platform')
plt.ylabel('age')
plt.title('Age Distribution by Platform')
plt.legend(title='Age')
plt.show()
```

```python
# Group by platform and calculate the average age that are using platform
age_by_platform = data.groupby('platform')['age'].mean()

# Plot the bar chart
age_by_platform.plot(kind='bar', color='skyblue')

# Add labels and title
plt.xlabel('Platform')
plt.ylabel('Average Age')
plt.title('Average Age by Platform')

# Show the plot
plt.show()
```

```python
# Scatter plot based on age and income
data.plot(kind='scatter', x='age', y='income', color='blue', alpha=0.7)

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Scatter Plot of Age vs. Income')

# Show the plot
plt.show()
```

```python
# Time spent on different platforms
time_by_platform = data.groupby('platform')['time_spent'].mean()

# Plot the bar chart
time_by_platform.plot(kind='bar', color='skyblue')

# Add labels and title
plt.xlabel('Platform')
plt.ylabel('Average Time Spent (hours)')
plt.title('Average Time Spent by Platform')

# Show the plot
plt.show()
```

```python
numeric_columns = ['age', 'income', 'time_spent', 'location', 'interests', 'profession', 'platform']
for column in numeric_columns:
    sns.histplot(data[column].dropna(), kde=True , color='blue')
    plt.title(f'Distribution of {column}')
    plt.show()
```
