```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\reddy\Downloads\crime_data.csv\crime_data.csv")
df
```

```python
#What is the overall trend in crime rates over time?
df['Date Rptd'] = pd.to_datetime(df['Date Rptd'])

# Resample data by month and count occurrences
monthly_crimes = df.resample('ME', on='Date Rptd').size()

# Plot
plt.figure(figsize=(12, 6))
monthly_crimes.plot()
plt.title('Monthly Crime Trends')
plt.xlabel('Month')
plt.ylabel('Number of Crimes')
plt.grid(True)
plt.show()
```

```python
#data of Only area name = Hollywood 
df[df['AREA NAME'] == "Hollywood"]
```

```python
#What is the distribution of different types of crimes?
# Count occurrences of each crime type
crime_counts = df['Crm Cd'].value_counts()

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x=crime_counts.index, y=crime_counts.values)
plt.title('Distribution of Crime Types')
plt.xlabel('Crime Type')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=90)
plt.show()
```

```python
#Which areas have the highest number of reported crimes?
# Count occurrences of crimes by area
area_counts = df['AREA NAME'].value_counts()

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x=area_counts.index, y=area_counts.values)
plt.title('Number of Crimes by Area')
plt.xlabel('Area')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=90)
plt.show()
```

```python
#What are the most common weapons used in crimes?
# Count occurrences of each weapon type
weapon_counts = df['Weapon Desc'].value_counts()

# Plot
plt.figure(figsize=(8, 8))
weapon_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Weapons Used')
plt.ylabel('')
plt.show()
```

```python
# What is the distribution of crime victims by age group?
# Bin age into groups
bins = [0, 18, 30, 40, 50, 60, 100]
labels = ['0-17', '18-29', '30-39', '40-49', '50-59', '60+']
df['age_group'] = pd.cut(df['Vict Age'], bins=bins, labels=labels)

# Plot
plt.figure(figsize=(12, 6))
sns.histplot(df['age_group'], discrete=True, stat='count')
plt.title('Distribution of Victims by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Victims')
plt.show()
```

```python
# Boxplot
under_18_data = df[df["Vict Age"] < 18]
plt.subplot(1, 2, 2)
sns.boxplot(x=under_18_data["Vict Age"], color='lightgreen')
plt.title('Boxplot of Victim Ages Under 18')
plt.xlabel('Age')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
# Count of each crime type
top_n = 15
top_crimes = df['Crm Cd Desc'].value_counts().nlargest(top_n)
sns.barplot(x=top_crimes.values, y=top_crimes.index)
plt.title(f'Top {top_n} Most Common Crime Types', fontsize=16)
plt.xlabel('Number of Incidents', fontsize=14)
plt.ylabel('Crime Type', fontsize=14)
plt.show()
```

```python
#plt.figure(figsize=(12, 6))
sns.histplot(df['Vict Age'].dropna(), bins=30, kde=True)
plt.title('Distribution of Victim Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
## Select numeric columns for correlation analysis
numeric_data = df.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
