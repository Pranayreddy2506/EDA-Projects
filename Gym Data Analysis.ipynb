```python

# Find the maximum weight class and the corresponding age
max_weight = dataset['Weight Class'].max()
max_age = dataset[dataset['Weight Class'] == max_weight]['Age'].values[0]

# Print the result
print(f"Maximum Weight Class: {max_weight}, Age: {max_age}")

# Visualization
plt.scatter(dataset['Age'], dataset['Weight Class'])
plt.scatter(max_age, max_weight, color='red', label='Max Weight')
plt.xlabel('Age')
plt.ylabel('Weight Class')
plt.title('Age vs Weight Class')
plt.legend()
plt.show()
```

```python
max_weight = dataset['Weight Class'].max()
max_age = dataset[dataset['Weight Class'] == max_weight]['Age'].values[0]

# Print the result
print(f"Maximum Weight Class: {max_weight}, Age: {max_age}")

# Visualization using Seaborn bar plot
sns.barplot(x='Age', y='Weight Class', data=dataset)
plt.bar(max_age, max_weight, color='red', label='Max Weight Class')
plt.xlabel('Age')
plt.ylabel('Weight Class')
plt.title('Age vs Weight Class')
plt.legend()
plt.show()
```

```python
# Visualization using Seaborn boxplot to compare lift types
#plt.figure(figsize=(10, 6))
sns.boxplot(x='Lift Type', y='Weight Class', data=dataset, palette='Set3',hue='Lift Type')
plt.xlabel('Lift Type')
plt.ylabel('Weight Class')
plt.title('Comparison of Weight Lifted by Lift Type')
plt.show()
```

```python
# Define age groups
dataset['age_group'] = pd.cut(dataset['Age'], bins=[10, 30, 60], labels=['10 to 30', '30 to 60'])

# Count lift types for each age group
counts = dataset.groupby(['age_group', 'Lift Type'], observed=False).size().reset_index(name='count')


# Visualization using Seaborn bar plot
sns.barplot(x='Lift Type', y='count', hue='age_group', data=counts, palette='Set2')
plt.xlabel('Lift Type')
plt.ylabel('Count')
plt.title('Count of Lift Types by Age Group')
plt.show()
```

```python
# Get the top 5 maximum amounts lifted
top_5_lifts = dataset.nlargest(5, 'Amount Lifted (kg)')
# Visualization using Seaborn bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Lifter Name', y='Amount Lifted (kg)', data=top_5_lifts, palette='viridis',hue='Lifter Name')
plt.xlabel('Lifter Name')
plt.ylabel('Amount Lifted (kg)')
plt.title('Top 5 Maximum Amounts Lifted')
plt.show()
```

```python
# Calculate the counts for each weight class
weight_class_counts = dataset['Lift Type'].value_counts()
# Plot the pie chart
plt.figure(figsize=(4, 4))
plt.pie(weight_class_counts, labels=weight_class_counts.index, autopct='%1.1f%%', colors=sns.color_palette('muted'), startangle=140)
plt.title('Distribution of Lift Type')
plt.show()
```

```python
sns.lineplot(x='Age', y='Amount Lifted (kg)', data=dataset, marker='o', color='b')
plt.title('Amount Lifted by Age')
plt.xlabel('Age')
plt.ylabel('Amount Lifted (kg)')
plt.grid(True)
plt.show()
```
